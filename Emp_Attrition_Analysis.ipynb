**Data Encoding:**


```python
data_attr["Attrition"] #target feature, 1-out of the company,0-still employed
```

```python
data_attr["Attrition"]=data_attr["Attrition"].replace(["Yes","No"],[1,0])
```

```python
data_attr["Attrition"]
```

```python
data_attr["Gender"]=data_attr["Gender"].replace(["Male","Female"],[1,0])
```

```python
data_attr["Gender"]
```

```python
data_attr["OverTime"]=data_attr["OverTime"].replace(["Yes","No"],[1,0])
```

```python
y=data_attr["Attrition"]#TARGET
```

```python
data_attr.drop(["Attrition"], axis=1, inplace=True)
```

```python
data_attr.shape
```

```python
features=["BusinessTravel","Department","EducationField","MaritalStatus","JobRole"]
Xnew=pd.get_dummies(data_attr[features])
```

```python
Xnew.columns
```

```python
data_attr.columns
```

```python
data_attr[["Age","DailyRate","DistanceFromHome","Education","EmployeeNumber",
           "EnvironmentSatisfaction","Gender","HourlyRate","JobInvolvement",
           "JobLevel","JobSatisfaction","MonthlyIncome","MonthlyRate","NumCompaniesWorked",
           "OverTime","PercentSalaryHike","PerformanceRating","PerformanceRating","StockOptionLevel",
           "TotalWorkingYears","TrainingTimesLastYear","WorkLifeBalance","YearsAtCompany","YearsInCurrentRole",
           "YearsSinceLastPromotion","YearsWithCurrManager"]]
```

```python
data_attr.reset_index(drop=True,inplace=True)
Xnew.reset_index(drop=True,inplace=True)
```

```python
X=pd.concat([data_attr[["Age","DailyRate","DistanceFromHome","Education","EmployeeNumber",
           "EnvironmentSatisfaction","Gender","HourlyRate","JobInvolvement",
           "JobLevel","JobSatisfaction","MonthlyIncome","MonthlyRate","NumCompaniesWorked",
           "OverTime","PercentSalaryHike","PerformanceRating","PerformanceRating","StockOptionLevel",
           "TotalWorkingYears","TrainingTimesLastYear","WorkLifeBalance","YearsAtCompany","YearsInCurrentRole",
           "YearsSinceLastPromotion","YearsWithCurrManager"]],Xnew],axis=1)
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
X_train.head()
```

```python
y_train[:10]
```

**Scaling**

```python
from sklearn.preprocessing import StandardScaler
```

```python
SS=StandardScaler()
```

```python
X_train=SS.fit_transform(X_train)
```

```python
X_test=SS.transform(X_test)
```

**Model and Feature Importance**

**LOGISTIC REGRESSION**

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg=LogisticRegression()
```

```python
logreg.fit(X_train,y_train)
```

```python
y_pred1= logreg.predict(X_test)
y_pred1
```

```python
from sklearn.metrics import  f1_score, recall_score, precision_score, accuracy_score, classification_report
```

```python
train_acc=logreg.score(X_train,y_train)
train_acc
```

```python
test_acc= accuracy_score(y_test,y_pred1)
test_acc
```

```python
print(classification_report(y_test,y_pred1))
```

```python
recal=recall_score(y_test,y_pred1)
prec=precision_score(y_test,y_pred1)
f1=f1_score(y_test,y_pred1)
print("Training Accuracy :", train_acc)
print("Testing Accuracy :", test_acc)
print("F1 Score :", f1)
print("Recall :", recal)
print("Precision :", prec)
```

```python
from sklearn.metrics import confusion_matrix
conf_mat= pd.DataFrame(data = confusion_matrix(y_test,y_pred1),
                           columns = ['Predicted:0', 'Predicted:1'],
                           index =['Actual:0', 'Actual:1'])
```

```python
plt.figure(figsize = (5, 3))
sns.heatmap(conf_mat, annot = True, fmt = 'd')
plt.show()
```

```python
coefficients = logreg.coef_[0]
```

```python
feature_importance = pd.DataFrame({'Feature': X.columns, 'Importance': np.abs(coefficients)})
feature_importance = feature_importance.sort_values('Importance', ascending=False)
```

```python
feature_importance.plot(x='Feature', y='Importance', kind='barh', figsize=(17,15))
```

So,major features for attrition based on this model are OverTime,YearsAtCompany,YearsInCurrentRole.
